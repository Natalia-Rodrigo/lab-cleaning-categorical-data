```python
def cont_discr_df(df: pd.DataFrame, num: int = 36):
    "This function takes as input a dataframe only with numerical variables and nÂºunique variables -> to discriminate between continuous and discrete variables, returning two different dataframes with each type of variable"
    cont_col_list = []
    discr_col_list = []
    
    for col in df:
        if len(df[col].unique()) > num:
            cont_col_list.append(col)
        else:
            discr_col_list.append(col)
    continuous_variables = df[cont_col_list].copy()
    discrete_variables = df[discr_col_list].copy()
    return continuous_variables, discrete_variables
```

```python
continuous_df, discrete_df = cont_discr_df(customer_df_num, 36)
display(continuous_df.head())
display(discrete_df.head())
```

```python
def barplot_discrete_variables(df: pd.DataFrame, figsize_height: int=16, figsize_width: int=16):
    """This functions takes as:
    Input: a df with discrete variables, and figsizes with a default value each of 16
    Output: bar plots for each column of the df"""
    for col in df:
        fig, ax = plt.subplots(figsize=(figsize_height,figsize_width))
        sns.barplot(data=df, x=col, y=col)
        plt.show()
```

```python
def histogram_continuous_variables(df: pd.DataFrame, figsize_height: int=16, figsize_width: int=16):
    """This functions takes as:
    Input: a df with continuous variables, and figsizes with a default value each of 16
    Output: histogram for each column of the df"""
    for col in df:
        fig, ax = plt.subplots(figsize=(figsize_height,figsize_width))
        sns.histplot(data=df, x=col)
        plt.show()
```

```python
def countplot_categorical_variables(df: pd.DataFrame, figsize_height: int=16, figsize_width: int=16):
    """This functions takes as:
    Input: a df with categorical variables, and figsizes with a default value each of 16
    Output: countplot for each column of the df sorted by frequency. Whenever the nÂº unique values > 6, bars will be placed in the y-axis"""
   
    for col in df:
        if len(df[col].unique()) >= 6:
            fig, ax = plt.subplots(figsize=(figsize_width, figsize_height))
            sns.countplot(data=df, y=col, order=df[col].value_counts().index, ax=ax)
            ax.set_title(f'Countplot for {col}')
        else:
            fig, ax = plt.subplots(figsize=(figsize_width, figsize_height))
            sns.countplot(data=df, x=col, order=df[col].value_counts().index, ax=ax)
            ax.set_title(f'Countplot for {col}')
            
        plt.show()
```

Define a function to remove the outliers of a numerical continuous column depending if a value is bigger or smaller than a given amount of standard deviations of the mean (thr=3).

```python
def remove_outliers_continuous(df: pd.DataFrame, thr: int = 3):
    """
    Remove outliers from a DataFrame with numerical continuous columns.
    Parameters:
    - df (pd.DataFrame): The input DataFrame.
    - thr (int, optional): Number of standard deviations from the mean to define the outlier range.
    Returns:
    - pd.DataFrame: A new DataFrame with outliers removed.
    """
    df1 = df.copy()
    for col in df1.columns:
        col_mean = df1[col].mean()
        col_std = df1[col].std()
        lower_limit = col_mean - thr * col_std
        upper_limit = col_mean + thr * col_std
        df1 = df1[(df1[col] >= lower_limit) & (df1[col] <= upper_limit)]
    return df1
```
